```python
# Default _corrupt_record column

from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('age', IntegerType(), True),
    StructField('_corrupt_record', StringType(), True) # stores corrupted records
])


df = spark.read.format('csv').options(header="true").schema(schema).load('/Volumes/spark_workouts/customerdb/customerfile/corruptdata/corrupted_csv_data.csv')

display(df)
```

```python
# User Defined column for storing corrupted records - set the user defined column name using the property "columnNameOfCorruptRecord"

from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField('id', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('age', IntegerType(), True),
    StructField('corrupted_record', StringType(), True) # User defined column name for storing corrupted records
])


df = spark.read.format('csv').options(header="true", columnNameOfCorruptRecord='corrupted_record').schema(schema).load('/Volumes/spark_workouts/customerdb/customerfile/corruptdata/corrupted_csv_data.csv')

display(df)
```
