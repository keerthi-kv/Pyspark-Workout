Read customer file and process it <br>
Download the custs.txt file from github url(https://github.com/azurede007/pycharmworkouts/blob/master/Data/custs.txt) <br>
Create volume called "customerfile" under the database "customerdb" under the catalog "spark_workouts" <br>
Create a folder called custdata in the volume customerfile and copy the downloaded file <br>
Create a folder under the home called "PySparkSQL-Excercise" <br>
Create a notebook called "UC01-custdataprocess" <br>
Read the custs.txt uploaded file with schema, filter only the profession = pilot <br>
Write the output in the volume folder path customerfile/pilotdata <br>

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Define schema
schema = StructType([
    StructField("custid", IntegerType(), True),
    StructField("fname", StringType(), True),
    StructField("lname", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("profession", StringType(), True)
])

df = (spark.read
    .option("header", "true")
    .schema(schema)
    .csv("/Volumes/spark_workouts/customerdb/customerfile/custdata/custs.txt"))
df1 = df.filter(df['profession']=="Pilot")
df1.display()
df1.write.option("header", "true").mode("overwrite").csv("/Volumes/spark_workouts/customerdb/customerfile/pilotdata")
```
